```python
import random

def generate_board(n):
    """
    Generates a random board configuration for the N-Queens problem.
    Each queen is placed in a different column, and the row is randomly chosen.
    """
    return [random.randint(0,n-1) for _ in range(n)]


def compute_hurestic(board):
    # count the number of attacking pairs of queens
    n = len(board)
    h=0
    for i in range(n):
        for j in range(i+1,n):
            if board[i]== board[j] or abs(i-j)== abs(board[i]-board[j]):
                h+=1
    return h


def get_neighbors(board):
    """
    Generates all possible neighbors of the current board configuration.
    A neighbor is generated by moving one queen to a different row in its column.
    """
    neighbours=[]
    n=len(board)
    for col in range (n):
        for row in range(n):
            if board[col] != row:
                new_board = list(board)
                new_board[col]=row
                neighbours.append(new_board)
    return neighbours


def hill_climbing(n):
    current_board = generate_board(n)
    current_huerestic=compute_hurestic(current_board)

    while True:
        neighbours = get_neighbors(current_board)
        best_neighbour= current_board
        best_hueristic=current_huerestic

        for neighbour in neighbours:
            neighbour_hueristic = compute_hurestic(neighbour)
            if(neighbour_hueristic< best_hueristic):
                best_neighbour=neighbour
                best_hueristic=neighbour_hueristic

        if best_hueristic == current_huerestic:
            break
        else:
            current_board = best_neighbour
            current_huerestic = best_hueristic


    return current_board, current_huerestic



def print_board(board):
    n = len(board)
    for row in range(n):
        line = ""
        for col in range(n):
            if board[col] == row:
                line += " Q "
            else:
                line += " . "
        print(line)
    print()  # Add a blank line after the board


solution_board, huerestic = hill_climbing(8)
if huerestic == 0:
    print("Solution found!")
else:
    print("No solution found.")

print()
print("Final Board Configuration: ", solution_board)
print_board(solution_board)
print("Final Huerestic Value: ", huerestic)




```
